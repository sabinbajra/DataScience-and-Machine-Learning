```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8])
y = np.array([3,10])
plt.plot(x,y,'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8,20,35])
y = np.array([3,10,6,12])
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8,20,35])
y = np.array([3,10,6,12])
plt.plot(x,y,marker='X')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8,20,35])
y = np.array([3,10,6,12])
plt.plot(x,y,'*--g',ms=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Rent", "Transport", "Study", "Others"]

plt.pie(y, labels = mylabels)
plt.show() 
```
